```python
# =============================================
# VISUAL ENHANCEMENT - FIXED VERSION
# Error-free code untuk semua versi pandas
# =============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

print("üé® CREATING VISUAL ENHANCEMENTS - FIXED VERSION")
print("=" * 60)

# Set style yang cantik
plt.style.use('default')  # Use default style yang lebih compatible
sns.set_palette("husl")

print("\nüìà 1. COMPREHENSIVE DATA OVERVIEW")
print("=" * 50)

# Pretty data display TANPA alpha parameter
display(HTML("<h3 style='color: #2E86C1;'>üìä Dataset Overview</h3>"))
print(f"Dataset Shape: {df.shape[0]} rows √ó {df.shape[1]} columns")

print("\nüìã First 10 Records:")
display(df.head(10).style.background_gradient(cmap='viridis'))  # Removed alpha

print("\nüìà Statistical Summary:")
display(df.describe().round(2).style.background_gradient(cmap='coolwarm'))  # Removed alpha

print("\nüìä 2. BEAUTIFUL VISUALIZATIONS")
print("=" * 50)

# Dashboard besar dengan 6 grafik
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('üè• Medical Insurance Analysis Dashboard', fontsize=18, fontweight='bold')

# Age distribution
axes[0,0].hist(df['age'], bins=8, alpha=0.7, color='skyblue', edgecolor='navy')
axes[0,0].set_title('üë• Age Distribution', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Age (years)')
axes[0,0].set_ylabel('Frequency')
axes[0,0].grid(True, alpha=0.3)

# BMI distribution dengan mean line
axes[0,1].hist(df['bmi'], bins=8, alpha=0.7, color='lightgreen', edgecolor='darkgreen')
axes[0,1].axvline(df['bmi'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {df["bmi"].mean():.1f}')
axes[0,1].set_title('‚öñÔ∏è BMI Distribution', fontsize=12, fontweight='bold')
axes[0,1].legend()

# Charges by smoker dengan value labels
smoker_charges = df.groupby('smoker')['charges'].mean()
bars = axes[0,2].bar(smoker_charges.index, smoker_charges.values,
                     color=['lightcoral', 'lightblue'], alpha=0.7, edgecolor='black')
axes[0,2].set_title('üö¨ Average Charges by Smoker', fontsize=12, fontweight='bold')
axes[0,2].set_ylabel('Average Charges ($)')

# Add value labels
for bar, value in zip(bars, smoker_charges.values):
    axes[0,2].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
                   f'${value:,.0f}', ha='center', va='bottom', fontweight='bold', fontsize=10)

# Gender pie chart
gender_counts = df['sex'].value_counts()
colors = ['#FF9999', '#66B2FF']
axes[1,0].pie(gender_counts.values, labels=gender_counts.index,
              autopct='%1.1f%%', colors=colors, startangle=90)
axes[1,0].set_title('üë´ Gender Distribution', fontsize=12, fontweight='bold')

# Region bar chart
region_counts = df['region'].value_counts()
bars = axes[1,1].bar(region_counts.index, region_counts.values,
                     color='orange', alpha=0.7, edgecolor='darkorange')
axes[1,1].set_title('üó∫Ô∏è Distribution by Region', fontsize=12, fontweight='bold')
axes[1,1].tick_params(axis='x', rotation=45)

# Children distribution
children_counts = df['children'].value_counts().sort_index()
bars = axes[1,2].bar(children_counts.index, children_counts.values,
                     color='gold', alpha=0.7, edgecolor='goldenrod')
axes[1,2].set_title('üë∂ Children Distribution', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

print("\nüìä 3. CORRELATION HEATMAP")
print("=" * 50)

# Create correlation matrix
df_corr = df.copy()
df_corr['sex_num'] = df_corr['sex'].map({'male': 1, 'female': 0})
df_corr['smoker_num'] = df_corr['smoker'].map({'yes': 1, 'no': 0})

numeric_cols = ['age', 'bmi', 'children', 'charges', 'sex_num', 'smoker_num']
correlation_matrix = df_corr[numeric_cols].corr()

# Beautiful heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlBu_r', center=0,
            square=True, linewidths=0.5, fmt='.3f')
plt.title('üîó Correlation Matrix - Insurance Factors', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nüìä 4. STYLED TABLES")
print("=" * 50)

# Risk categorization table
df['risk_category'] = pd.cut(df['charges'],
                            bins=[0, 5000, 15000, float('inf')],
                            labels=['Low Risk', 'Medium Risk', 'High Risk'])

display(HTML("<h3 style='color: #E74C3C;'>üéØ Risk Categories Analysis</h3>"))
risk_table = df.groupby('risk_category').agg({
    'age': 'mean',
    'bmi': 'mean',
    'charges': ['mean', 'count']
}).round(2)
display(risk_table.style.background_gradient(cmap='Reds'))  # Removed alpha

# Demographic breakdown
display(HTML("<h3 style='color: #8E44AD;'>üë• Demographic Breakdown</h3>"))
demo_table = df.groupby(['sex', 'smoker']).agg({
    'charges': ['mean', 'count'],
    'age': 'mean',
    'bmi': 'mean'
}).round(2)
display(demo_table.style.background_gradient(cmap='viridis'))  # Removed alpha

# Key insights table
display(HTML("<h3 style='color: #D35400;'>üéØ Key Business Insights</h3>"))
insights_data = {
    'Metric': ['Total Records', 'Average Age', 'Average BMI', 'Average Charges', 'Smoker %'],
    'Value': [f"{len(df)}", f"{df['age'].mean():.1f}", f"{df['bmi'].mean():.1f}",
              f"${df['charges'].mean():,.0f}", f"{(df['smoker'] == 'yes').mean()*100:.1f}%"],
    'Insight': ['Good sample', 'Middle-aged', 'Overweight', 'Moderate costs', 'Minority smokers']
}
insights_df = pd.DataFrame(insights_data)
display(insights_df.style.background_gradient(cmap='autumn'))
```
